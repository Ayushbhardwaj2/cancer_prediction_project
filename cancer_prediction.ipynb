```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report


cancer = pd.read_csv('Cancer.csv')


y = cancer['diagnosis']
X = cancer.drop(['id', 'diagnosis', 'Unnamed: 32'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=2529)


model = LogisticRegression(max_iter=5000)


model.fit(X_train, y_train)

y_pred = model.predict(X_test)

conf_matrix = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

# Output results
print("Confusion Matrix:\n", conf_matrix)
print("Accuracy Score:", accuracy)
print("Classification Report:\n", class_report)


def predict_cancer(input_data):
    
    input_df = pd.DataFrame(input_data)
    
    
    predictions = model.predict(input_df)
    
    return predictions


new_patients = [
    {
        'radius_mean': 14.5,
        'texture_mean': 20.3,
        'perimeter_mean': 90.0,
        'area_mean': 600.0,
        'smoothness_mean': 0.1,
        'compactness_mean': 0.05,
        'concavity_mean': 0.02,
        'concave points_mean': 0.03,
        'symmetry_mean': 0.2,
        'fractal_dimension_mean': 0.06,
        'radius_se': 1.2,
        'texture_se': 1.0,
        'perimeter_se': 5.0,
        'area_se': 30.0,
        'smoothness_se': 0.005,
        'compactness_se': 0.01,
        'concavity_se': 0.01,
        'concave points_se': 0.01,
        'symmetry_se': 0.02,
        'fractal_dimension_se': 0.01,
        'radius_worst': 15.0,
        'texture_worst': 22.0,
        'perimeter_worst': 95.0,
        'area_worst': 650.0,
        'smoothness_worst': 0.12,
        'compactness_worst': 0.06,
        'concavity_worst': 0.03,
        'concave points_worst': 0.04,
        'symmetry_worst': 0.25,
        'fractal_dimension_worst': 0.07
    },
    {
        'radius_mean': 15.0,
        'texture_mean': 19.5,
        'perimeter_mean': 95.0,
        'area_mean': 650.0,
        'smoothness_mean': 0.12,
        'compactness_mean': 0.06,
        'concavity_mean': 0.03,
        'concave points_mean': 0.04,
        'symmetry_mean': 0.25,
        'fractal_dimension_mean': 0.07,
        'radius_se': 1.3,
        'texture_se': 1.1,
        'perimeter_se': 5.5,
        'area_se': 32.0,
        'smoothness_se': 0.006,
        'compactness_se': 0.02,
        'concavity_se': 0.02,
        'concave points_se': 0.02,
        'symmetry_se': 0.03,
        'fractal_dimension_se': 0.02,
        'radius_worst': 16.0,
        'texture_worst': 23.0,
        'perimeter_worst': 100.0,
        'area_worst': 700.0,
        'smoothness_worst': 0.13,
        'compactness_worst': 0.07,
        'concavity_worst': 0.04,
        'concave points_worst': 0.05,
        'symmetry_worst': 0.26,
        'fractal_dimension_worst': 0.08
    },
    
]

predictions = predict_cancer(new_patients)


for i, prediction in enumerate(predictions):
    print(f"Patient {i+1}: {'Malignant' if prediction == 'M' else 'Rachin'}")
```
